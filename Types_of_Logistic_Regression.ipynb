*** Simple Logistic Regression ***

```python
### Import Necessary Libraries
import pandas as pd
import numpy as np
```

```python
# Load the dataset from GitHub URL
url = "https://raw.githubusercontent.com/KenDaupsey/Types-of-Logistic-Regression/main/hsb2%7Edata.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
##### Convert Continous Variable "math" Into Binary Variable
#import pandas as pd

# Assuming df is your DataFrame
# Define the ranges for categories
category_ranges = [0, 49, 100]  # Adjust the ranges as needed

# Define the corresponding labels for categories
category_labels = [0, 1]

# Create the 'math2' column using the cut function
df['math2'] = pd.cut(df['math'], bins=category_ranges, labels=category_labels, include_lowest=True)

# Display the first few rows to verify the changes
print(df[['math', 'math2']].head())
```

```python
df.head()
```

```python
# Logistic Regression Procedure
import statsmodels.api as sm

X = df['write']  # Independent variable
y = df['math2']   # Dependent variable

# Add a constant term to the independent variable
X = sm.add_constant(X)

# Fit simple logistic regression model
logit_model = sm.Logit(y, X)
result = logit_model.fit()

# Print the summary
print(result.summary())
```

*** Ordered logistic regression ***

```python
### Import Necessaary Libraries
import numpy as np
import pandas as pd
```

```python
# Load the dataset from GitHub URL
url = "https://raw.githubusercontent.com/KenDaupsey/Types-of-Logistic-Regression/main/hsb2%7Edata.csv"
df = pd.read_csv(url)
df.head()
```

```python
df.columns
```

```python
## Creating New “write3” from ”write”
import pandas as pd
import statsmodels.api as sm

# Assuming df is your DataFrame
# Generate write3
df['write3'] = 1
df.loc[(df['write'] >= 49) & (df['write'] <= 57), 'write3'] = 2
df.loc[(df['write'] >= 58) & (df['write'] <= 70), 'write3'] = 3

df.head()
```

```python
df.columns
```

```python
df['write3'].value_counts()
```

```python
# Perform multinomial logistic regression
model = sm.MNLogit.from_formula('write3 ~ female + read + socst', data=df)
result = model.fit()

# Print the summary
print(result.summary())
```

*** Another Ordered Regression Procedure ***

```python
### Import Necessary Libraries
import pandas as pd
import numpy as np
```

```python
### Load Dataset
url= "https://raw.githubusercontent.com/KenDaupsey/Types-of-Logistic-Regression/main/MultionmialRegression%7EDataset.csv"
df = pd.read_csv(url)
df.head()
```

```python
### Import Library for Multinomial Logistic Regression Procedure
from statsmodels.miscmodels.ordinal_model import OrderedModel
```

```python
df.describe().T
```

```python
df.dtypes
```

```python
# Convert 'female' column to categorical dtype
df['female'] = df['female'].astype('category')

# Now you can use the .cat accessor to access the category codes
df['female'] = df['female'].cat.codes

# Check the data types
print(df.dtypes)
```

```python
# Convert input data to numpy arrays
endog = np.asarray(df['cid'])
exog = np.asarray(df[['female', 'write',  'read',  'math']])

# Fit the ordered probit model
mod_probit = OrderedModel(endog, exog, distr='probit')
res_probit = mod_probit.fit(method='bfgs')

# Fit the ordered logistic regression model
mod_logit = OrderedModel(endog, exog, distr='logit')
res_logit = mod_logit.fit(method='bfgs')

# Compare the model summaries
print(res_probit.summary())
print(res_logit.summary())
```

```python
### Model Performance Evaluation Using Information Criteria
# Evaluate Model Performance using AIC and BIC
print("AIC - Ordered Probit Model:", res_probit.aic)
print("BIC - Ordered Probit Model:", res_probit.bic)

print("AIC - Ordered Logistic Regression Model:", res_logit.aic)
print("BIC - Ordered Logistic Regression Model:", res_logit.bic)
```

```python
### Calculate the differences in AIC and BIC between the ordered logistic regression and ordered probit models
delta_AIC = res_logit.aic - res_probit.aic
delta_BIC = res_logit.bic - res_probit.bic
print("ΔAIC:", delta_AIC)
print("ΔBIC:", delta_BIC)
```

*** Multinomial Logistic Regression (MNlogit) ***

```python
### Import Necessary Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### Load Dataset and Examine first 5 Rows
url= "https://raw.githubusercontent.com/KenDaupsey/Types-of-Logistic-Regression/main/multiclassification%7Ecustomer%7Edata.csv"

df = pd.read_csv(url)
df.head()
```

```python
### Encoding Categorical Variables
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Convert categorical variables to integers
df['Education_Level'] = label_encoder.fit_transform(df['Education_Level'])
df['Marital_Status'] = label_encoder.fit_transform(df['Marital_Status'])
df['Income_Category'] = label_encoder.fit_transform(df['Income_Category'])
df['Health_Status'] = label_encoder.fit_transform(df['Health_Status'])

# Display the modified dataframe
print(df.head())
```

```python
df.dtypes
```

```python
df['Health_Status'].value_counts()
```

```python
### The Multinomial Logistic Regression Procedure
# Extract independent and dependent variables
X = df[['Age', 'Education_Level', 'Marital_Status', 'Credit_Score',
       'Annual_Income', 'Savings_Amount', 'Income_Category',]]
y = df['Health_Status']

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit multinomial logistic regression model
model = sm.MNLogit(y, X)
result = model.fit()

# Print the summary
print(result.summary())
```

```python
### test for proportional odds
import statsmodels.api as sm

# Specify the formula for proportional odds logistic regression
formula = 'Health_Status ~ Age + Education_Level + Marital_Status + Credit_Score + Annual_Income + Savings_Amount'
# Fit the proportional odds logistic regression model
model = sm.OLS.from_formula(formula, data=df).fit()

# Print the summary
print(model.summary())
```
